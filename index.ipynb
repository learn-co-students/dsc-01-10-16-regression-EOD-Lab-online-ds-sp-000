### Let's do Regression 

Right here is the real deal. Let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). We shall do this is a loop and in every iteration, we shall pick one of the independent variables  perform following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`.
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
# Your code here
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats
import statsmodels.stats.api as sms



results = [['ind_var', 'r_squared', 'intercept', 'slope', 'p-value', 'normality (JB)' ]]
for idx, val in enumerate(['crim', 'dis', 'rm', 'zn', 'age']):
    print ("Boston Housing DataSet - Regression Analysis and Diagnostics for formula: medv~" + val)
    print ("-------------------------------------------------------------------------------------")

    f = 'medv~' + val
 
    model = ols(formula=f, data=newdf).fit()
    
    X_new = pd.DataFrame({val: [newdf[val].min(), newdf[val].max()]});
    preds = model.predict(X_new)
    newdf.plot(kind='scatter', x=val, y='medv');
    plt.plot(X_new, preds, c='red', linewidth=2);
    plt.show()
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, val, fig=fig)
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
    fig.show()
    
    results.append([val, model.rsquared, model.params[0], model.params[1], model.pvalues[1], sms.jarque_bera(model.resid)[0]])
    input("Press Enter to continue...")
```

```python
pd.DataFrame(results)
```

```python
#Your obervations here 
# There is a negative relationship between crime and median value of owner occupied homes (e.g., appears
# as though as median value increases crime decreases).
# Crime scatterplot shows no sign of heteroscadasticity.
# Residuals for crime QQ Plots don't show much sign of normality.
# Crime R-squared is very low therefore a low percent of the variations in median value can be explained by crime.

# A positive relationship is present between dis and medv.
# Scatterplot shows some sign of heteroscadasticity between dis and medv.
# Residuals don't show much sign of normality.
# R-squared is extremely low therefore very little percent of the variations in median value can be explained by dis.

# There appears to be a very positive relationship between rm and medv.
# Residuals have no cone shape, therefore no sign of heteroscadasticity between rm and medv.
# R-squared is much higher than the other features.  It's value of 48% implies that 48% of the variations
# in medv can be explained by the number of rooms per dwelling.

# There is a lot of variance in the zn scatterplot and shows no sign of normality.
# Residuals show no sign of heteroscadasticity between zn and medv.
# R-squared is very low therefore a low percent of the variations in the median value can be explained by zn.

# It appears that age and median value have a negative relationship while median value is greater the younger the age.
# Some cone shape to residuals, therefore signs of heteroscadsticity are present.
# Based on QQ plot residuals for age do not show signs of normal distribution.
```
